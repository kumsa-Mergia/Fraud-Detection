# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from pathlib import Path
sns.set(style="whitegrid")
```

```python
# Load Cleaned Credit Card Data
data_path = Path("../../data/processed/cleaned_balanced_fraud_data.csv")
df = pd.read_csv(data_path)
df.head()
```

# Class Distribution

```python
sns.countplot(x='class', data=df)
plt.title("Balanced Class Distribution")
plt.show()
```

# Feature Distribution (Numeric)

```python
num_features = ['purchase_value', 'age', 'hour_of_day', 'day_of_week', 'time_since_signup']

df[num_features].hist(bins=30, figsize=(12, 8), color="#69b3a2", edgecolor="black")
plt.suptitle("Distributions of Scaled Numeric Features", fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
