# Notebook: 02 - EDA on Credit Card Data

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from pathlib import Path
sns.set(style="whitegrid")
```

```python
# Load Cleaned Credit Card Data
data_path = Path("../../data/processed/cleaned_credit_data.csv")
df = pd.read_csv(data_path)
df.head()
```

## Univariate Analysis

```python
# Set overall figure aesthetics
plt.figure(figsize=(15, 10))

# Plot histograms for all numeric columns
ax = df.hist(
    bins=30, 
    figsize=(15, 10), 
    color="#69b3a2", 
    edgecolor="black", 
    grid=False
)

# Customize each subplot
for a in ax.ravel():
    a.set_title(a.get_title(), fontsize=12, fontweight='bold')
    a.set_xlabel(a.get_xlabel(), fontsize=10)
    a.set_ylabel("Frequency", fontsize=10)
    a.grid(axis='y', linestyle='--', alpha=0.5)

plt.suptitle("Distributions of Numerical Features", fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.96])  # leave space for suptitle
plt.show()
```

## Bivariate Analysis

```python
# Compute correlation matrix for numeric columns
corr = df.corr(numeric_only=True)

# Set up the matplotlib figure
plt.figure(figsize=(17, 17))

sns.heatmap(
    corr, 
    annot=True, 
    fmt=".2f", 
    cmap="coolwarm", 
    center=0,
    square=True,
    linewidths=0.5, 
    linecolor='gray',
    cbar_kws={"shrink": 0.8}
)

plt.title("Feature Correlation Heatmap", fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)
plt.tight_layout()
plt.show()
```
